```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0 , 10 , 0.1)
y = np.sin(x)


plt.plot(x , y)


plt.title("sine graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show
```

```python
#data

x = np.random.rand(20)
y = np.random.rand(20)
colors = np.random.rand(20)
sizes = np.random.randint(20 , 80 , 20)

#creating a graph

plt.scatter(x , y , c=colors  , s=sizes)


#Determie chart propeties

plt.title("point chart")
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.show()
```

```python
#data
data = np.random.randn(500)

#Graph Creation
plt.hist(data , bins = 10)

#Determine graph properties 

plt.title("Histogram")
plt.xlabel("value")
plt.ylabel("Frequency")

plt.show()
```

```python
x = ['A' , 'B' , 'C' , 'D' , 'E']
y = np.random.randint(1, 10 , 5)


plt.bar(x,y)


plt.title("column chart")
plt.xlabel("Category")
plt.ylabel("Value")


plt.show()
```

```python
sizes = [32,65,6,34,52,4,34,34]

plt.pie(sizes)


plt.title("pie chart")


plt.show()
```
